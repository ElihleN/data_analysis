## Feature engineering


```python
#first convert to datetime values
df['Ship Date'] = pd.to_datetime(df['Ship Date'],dayfirst = True)
df['Order Date'] = pd.to_datetime(df['Order Date'],dayfirst = True)

# Extract features (columns) by year, month, and quarter.
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
# df['Quarter'] = df['Order Date'].dt.quarter
df['Week'] = df['Order Date'].dt.isocalendar().week


df['Quarter'] = 'Q' + df['Order Date'].dt.quarter.astype(str)


```

#### Aggregation and Grouping


```python
# # Aggregate calculations across various time dimensions

# Aggregate sales data by different time dimensions
monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

quarterly_sales = df.groupby(['Year', 'Quarter'])['Sales'].sum().reset_index()
print(quarterly_sales)

yearly_sales = df.groupby('Year')['Sales'].sum().reset_index()

# Create a time series index
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.set_index('Order Date', inplace=True)

# Daily Sales Data
daily_sales = df['Sales'].resample('D').sum()

# Monthly Sales Data
monthly_sales_ts = df['Sales'].resample('ME').sum()
```

### Time Series Decomposition

```python
# Time Series Decomposition and Visualization
from statsmodels.tsa.seasonal import seasonal_decompose
# Ensure there are no missing values
monthly_sales_ts = monthly_sales_ts.fillna(monthly_sales_ts.mean())

## Perform decomposition
# decomposition = seasonal_decompose(monthly_sales_ts, model='additive', period=12).  # Assuming - Constant seasonal amplitude and  Linear combination of components
decomposition = seasonal_decompose(monthly_sales_ts, model='multiplicative',  period=12)  # Assuming - proportional seasonality (seasonal swings grow with sales level)

plt.figure(figsize=(15, 12))
plt.subplot(4, 1, 1)
decomposition.observed.plot()
plt.title('Observed value')

plt.subplot(4, 1, 2)
decomposition.trend.plot()
plt.title('Trend')

plt.subplot(4, 1, 3)
decomposition.seasonal.plot()
plt.title('Seasonal')

plt.subplot(4, 1, 4)
decomposition.resid.plot()
plt.title('Residual')

plt.tight_layout()
plt.show()
```



*   Based on the Observed series, sales are growing over time.
    - Seasonality is pronounced and repeats annully
    -  Higher volatility in later years is largely explained by the growing trend rather than random noise.
*   From the trend we observe  a strong, steady growth which is accelerated during 2026-2017.
- This means the company is experiencing sustained expansion.
- Suggestion: Market expansion

* Residual component  is centered around zero with no visible systematic pattern.



```python
# Rolling (moving) averages over a monthly sales time series.

# Calculate 3-month and 6-month moving averages.
rolling_3m = monthly_sales_ts.rolling(window=3).mean()
rolling_6m = monthly_sales_ts.rolling(window=6).mean()

plt.figure(figsize=(12, 6))
monthly_sales_ts.plot(label='Original data')
rolling_3m.plot(label='Three-month moving average', color='red')
rolling_6m.plot(label='Six-month moving average', color='green')
plt.title('Sales volume and moving average')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

. There is a **clear upward long-term growth trend**.

. Sales exhibit **high short-term volatility**.

. There appears to be **annual seasonality**.


```python
# Autocorrelation analysis.
from statsmodels.graphics.tsaplots import plot_acf

plt.figure(figsize=(12, 6))
plot_acf(monthly_sales_ts, lags=47)
plt.title('Sales auto-correlation')
plt.xlabel('Lag order')
plt.ylabel('Autocorrelation coefficient')
plt.show()
```

As shown in the graph:


* Lag 1 is positive and significant

  - Sales this month are correlated with last month.
This indicates short-term persistence.

* Strong spike at Lag 12 (~0.55), confirming Strong annual seasonality; Monthly data with 12-period cycle and recurring yearly demand pattern







# Logistics Method Analysis: .

```python
# The percentage of orders and sales revenue for each method.
ship_mode_analysis = df.groupby('Ship Mode').agg(
    Order_Count=('Order ID', 'nunique'),
    Total_Sales=('Sales', 'sum'),
    Avg_Sales=('Sales', 'mean')
    # Avg_Process_Time=('Process Time', 'mean')
).reset_index()

## Calculate percentage
ship_mode_analysis['Order_Pct'] = ship_mode_analysis['Order_Count'] / ship_mode_analysis['Order_Count'].sum() * 100
ship_mode_analysis['Sales_Pct'] = ship_mode_analysis['Total_Sales'] / ship_mode_analysis['Total_Sales'].sum() * 100

print(ship_mode_analysis)
```

```python
# Visualization

plt.figure(figsize=(15, 10))

# Order distribution by mode of transportation

plt.subplot(2, 2, 1)

sns.barplot(data=ship_mode_analysis, x='Ship Mode', y='Order_Pct')

plt.title('The proportion of orders by each mode of transportation')

plt.ylabel('Order proportion (%)')

# Sales distribution by mode of transportation

plt.subplot(2, 2, 2)

sns.barplot(data=ship_mode_analysis, x='Ship Mode', y='Sales_Pct')

plt.title('The proportion of sales by each mode of transportation')

plt.ylabel('Sales proportion (%)')

# Transportation mode and average order value

plt.subplot(2, 2, 3)

sns.barplot(data=ship_mode_analysis, x='Ship Mode', y='Avg_Sales')
plt.title('Average order value of each mode of transportation')
plt.ylabel('Average sales volume($)')

plt.tight_layout()
plt.show()
```

###Trends in transportation modes.


```python

ship_mode_trend = df.groupby(['Year', 'Ship Mode'])['Sales'].sum().unstack()

ship_mode_trend.plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Annual sales trends of each mode of transportation')
plt.ylabel('Sales($)')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```

* Shipping Mode Distribution: Displays the order and sales percentages for different shipping modes (Standard Class, Second Class, etc.). The chart shows that standard delivery accounts for 60% of orders, while same-day delivery accounts for the fewest. The sales percentage and order volume percentage for each shipping mode are generally consistent, with standard delivery also accounting for nearly 60% of sales. The average order value for each shipping mode is similar, ranging from $225 to $240.

* Shipping Efficiency: Analyzes the average processing time for different shipping modes. The chart shows that standard delivery has the longest processing time, averaging 5 days; same-day delivery has the shortest processing time, less than 1 day; while priority delivery and next-day delivery both take approximately 2-3 days.

* Trend Changes: Observes the trends of each shipping mode over time. The chart shows that standard delivery sales change most significantly with year and time, showing an overall upward trend, and is expected to continue rising in the future. The other three shipping modes show slight upward trends, but the growth rates are small.

```python
# Prepare data for map visualizations (such as using plotly or geopandas), where state codes are often used for geographic mapping.

# Create a DataFrame mapping state names and state codes

state_df = pd.DataFrame({'State': state, 'State Code': state_code})

# Calculate total sales by state and sort in descending order

sales = df.groupby("State")["Sales"].sum().sort_values(ascending=False)

# Convert state names from index to regular columns (the DataFrame now has only two columns: State and Sales)

sales = sales.reset_index()

# Sort by state names in ascending alphabetical order (the original code contradicts the previous descending sort; requirements need confirmation)

sales = sales.sort_values('State', ascending=True)

# Reset the index and delete the old index column (the reset_index/drop operations in the original code can be combined into one step)

sales = sales.reset_index(drop=True)

# Insert a state code column (ensure the state order in state_df matches that of sales)

sales['State Code'] = state_df.set_index('State')['State Code'].reindex(sales['State']).values
```

```python
# # Sales distribution heatmap, suitable for displaying business performance in different regions.
import plotly.graph_objects as go

# # Prepare floating text: Create a new column named "text" and copy the state name data (to display the full state name when hovering over the map).
sales['text'] = sales['State'] + '<br>' + 'sales: $' + sales['Sales'].astype(str)

# Create a blue gradient color scheme
custom_blue_scale = [
    [0.0, '#f7fbff'],   #
    [0.2, '#deebf7'],
    [0.4, '#c6dbef'],
    [0.6, '#9ecae1'],
    [0.8, '#6baed6'],
    [1.0, '#2171b5']
]

# #  Create an isotropic map.
fig = go.Figure(data=go.Choropleth(
    locations=sales['State Code'],  # # State abbreviation code
    text=sales['text'],             # Text displayed on hover
    z=sales['Sales'].astype(float), ## Color mapping value
    locationmode='USA-states',      ## US state-level map
    colorscale=custom_blue_scale,   # # Custom blue gradient
    colorbar_title="sales(＄)",
    marker_line_color='#2171b5',
    marker_line_width=0.7           # State boundary line width
))

# # Set map layout
fig.update_layout(
    title_text='<b>Sales distribution in each state of the United States</b>',
    title_font=dict(size=24, family='Arial', color='#333333'),
    title_x=0.5,
    geo=dict(
        scope='usa',
        projection=go.layout.geo.Projection(type='albers usa'),
        showlakes=True,
        lakecolor='rgb(255, 255, 255)',
        landcolor='rgb(240, 240, 240)',
        subunitcolor='#2171b5'
    ),
    margin=dict(l=0, r=0, t=40, b=0),
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)
# Add custom hover effect
fig.update_traces(
    hovertemplate='<b>%{text}</b><extra></extra>',
    hoverlabel=dict(
        bgcolor='white',
        font_size=14,
        font_family='Arial'
    )
)
fig.show()
```
